```python
# This file contains examples on how to create different domains.
# First import what basic domains are needed, here:
from neural_diff_eq.problem.domain.domain2D import Rectangle, Circle
# See documentation of the Domain-class for all possible objects
```

```python
# Create instances of the domains. 
R = Rectangle([0, 0], [1, 0], [0, 1])
C = Circle([0, 0], 0.5)
# If only simple domains are needed, everything is already done and could be 
# connected to a variable. E.g. x = Variable(name='x', domain=R,...)
```

```python
# For complex domains there exists different operations to manipulate 
# the basic domain. For example lets cut the circle from the rectangle
from neural_diff_eq.problem.domain.domain_operations import Cut
cut = Cut(R, C)
# create some points on the boundary to see if everything worked
import matplotlib.pyplot as plt
b = cut.sample_boundary(50, type='grid')
p = plt.scatter(b[:, 0], b[:, 1])
```

```python
# If there already exists a file, containing some 3D object, 
# that should be the domain. We can load it with the Polygon3D class:
from neural_diff_eq.problem.domain.domain3D import Polygon3D
poly3D = Polygon3D(file_name='cornerplate.stl', file_type='stl')
# to look at the object:
# poly3D.mesh.show()
```

```python
# This is a now 3 dimensional domain. If only a 2D representation of 
# the object is needed, the polygon can either be sliced with a plane
# or projected on a plane.
# Lets project it onto the xy-plane:
poly2D = poly3D.project_on_plane(plane_origin=[0, 0, 0], 
                                 plane_normal=[0, 0, 1])
# This returns a Polygon2D-object
b = poly2D.sample_boundary(50, type='grid')
p = plt.scatter(b[:, 0], b[:, 1])
```
